```python
#pandas is a library and it process the datas (spreadsheet table format ) 
import pandas as pd
```

```python
#the salary data is saving in dataset variable to open the data file
dataset=pd.read_csv("Salary_Data.csv")
```

```python
#save the coloumn yearsexperience as independent variable to predict yearexperience
independent=dataset[["YearsExperience"]]
```

```python
#save the column salary as dependent variable to predict salary
dependent=dataset[["Salary"]]
```

```python
#sklearn is a library and it contain mathematical datas
#import the split function to split datasets into training and testing data
from sklearn.model_selection import train_test_split
```

```python
#here the datasets split into the ratio of 1/3 70%training and 30% testing data
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
#import linearregression because of straight line model and here the linear regression is a class 
from sklearn.linear_model import LinearRegression
```

```python
#saving linearregression into regressor variable 
#model intilizing
regressor=LinearRegression()
```

```python
# fitting the model on the training data 
regressor.fit(x_train,y_train)
```

```python
#coefficent is a function and here it predict the how much salary increased per year using learned model  
weight=regressor.coef_
weight
```

```python
#here the intercept is predict the origin(bias) salary when yearsof experience is 0
bias=regressor.intercept_
bias
```

```python
#used to predict the input test data with training model and saved in variable y pred(predicted data)
y_pred=regressor.predict(x_test)
```

```python
#import r2 metric to evalute the how well the model perform
from sklearn.metrics import r2_score
```

```python
#calculate the r2 score between actual salary data with predicted salary data
r_score=r2_score(y_test,y_pred)
```

```python
#import pickle to save the trained model 
import pickle
```

```python
#named the trained model as finalized model and saving in the filename variable
filename="finalized_model_linear.sav"
```

```python
#saving trained model in pickle 
pickle.dump(regressor,open(filename,'wb'))
```

```python
#testing the trained model that it predict properly
loaded_model=pickle.load(open("finalized_model_linear.sav",'rb'))
```

```python
#predicting the salary in 15 years of experiences in assigned result variable
result=loaded_model.predict([[15]])
```
